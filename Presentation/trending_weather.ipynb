```python
import pandas as pd
```

```python
df = pd.read_csv('US Precipitation Historic.csv')
df = df[['Unnamed: 0', 'Washington', 'Oregon', 'California', 'New York']]
df = df.rename(columns={'Unnamed: 0': "Year"})
df = df[df.Year > 1990]
df['us_mean'] = df.iloc[:, 1:5].mean(axis=1)
us_mean_df = df[['Year', 'us_mean']]
us_mean_df
```

```python
df = pd.read_csv('France Precipitation Historic.csv')
df = df[['Unnamed: 0', 'Alsace', 'Aquitaine', 'Bourgogne', 'Champagne-Ardenne']]
df = df.rename(columns={'Unnamed: 0': "Year"})
df = df[df.Year > 1990]
df['fr_mean'] = df.iloc[:, 1:5].mean(axis=1)
fr_mean_df = df[['Year', 'fr_mean']]
fr_mean_df
```

```python
df = pd.read_csv('Italy Precipitation Historic.csv')
df = df[['Unnamed: 0', 'Piemonte', 'Sicilia', 'Toscana', 'Veneto']]
df = df.rename(columns={'Unnamed: 0': "Year"})
df = df[df.Year > 1990]
df['it_mean'] = df.iloc[:, 1:5].mean(axis=1)
it_mean_df = df[['Year', 'it_mean']]
it_mean_df
```

```python
df = pd.read_csv('Spain Precipitation Historic.csv')
df = df[['Unnamed: 0', 'Cantabria']]
df = df.rename(columns={'Unnamed: 0': "Year"})
df = df[df.Year > 1990]
df['sp_mean'] = df.iloc[:, 1:5].mean(axis=1)
sp_mean_df = df[['Year', 'sp_mean']]
sp_mean_df
```

```python
from functools import reduce
reduce(lambda x,y: pd.merge(x,y, on='Year', how='outer'), [us_mean_df, fr_mean_df, it_mean_df, sp_mean_df])
```
