```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read projected precipitation data in CSV to a DataFrame
file_path1 = "historical_weather.csv"
historical_df = pd.read_csv(file_path1)
historical_df.head()
```

```python
historical2_df = historical_df[historical_df['Year'] > 1990]
historical2_df
```

```python
historical2_df.set_index('Year', inplace=True)
historical2_df.groupby('Province')['Historical Precip'].plot(legend=True)
```
